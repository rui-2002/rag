```python
import os

sample_text = {
    "../data/test_files/sample.txt": """This is a sample text file for testing.
Another line of text for testing. This file is used to demonstrate the Document data structure in LangChain."""
,

    "../data/test_files/sample2.txt": """This is another sample text file for testing.
It contains multiple lines of text to provide more content for testing the Document data structure in LangChain"""
,
"../data/test_files/sample3.txt": """This is yet another sample text file for testing.
It contains multiple lines of text to provide more content for testing the Document data structure in LangChain and to demonstrate how to create multiple documents from a directory."""}



for filepath, content in sample_text.items():
    os.makedirs(os.path.dirname(filepath), exist_ok=True)  # Create folder if not exists
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)

print("Sample text file created successfully.")
```

```python
### TextLoader
from langchain_community.document_loaders import TextLoader


loader=TextLoader("../data/test_files/sample.txt",encoding="utf-8")
documents=loader.load()
documents
```

```python
### Directory loader
from langchain_community.document_loaders import DirectoryLoader    

dir_loader=DirectoryLoader("../data/test_files",
                            glob="*.txt",
                            loader_cls=TextLoader,
                            loader_kwargs={"encoding":"utf-8"}, 
                            show_progress=False)
documents=dir_loader.load()
documents
```

```python
## Pdf loader

from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader

## load all the text files from the directory

dir_loader=DirectoryLoader("../data/pdf",
                           glob="**/*.pdf",## pattern to match files
                           loader_cls=PyMuPDFLoader, ## loader class to use
                           show_progress=False)

documents=dir_loader.load()
documents
```
